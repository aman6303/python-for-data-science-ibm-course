```python
import requests
import pandas as pd  # Optional, if you want to put it in a DataFrame later

# 1. Define the URL for 10 users (same as generate_users(10))
url = 'https://randomuser.me/api/?results=10'

# 2. Set a User-Agent header to mimic a web browser
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36'
}

try:
    # 3. Make the request
    response = requests.get(url, headers=headers)
    
    # 4. Check if successful
    if response.status_code == 200:
        data = response.json()
        users = data['results']
        
        # Example: Print the first user's name to verify
        print(f"Successfully fetched {len(users)} users.")
        print(f"First user: {users[0]['name']['first']} {users[0]['name']['last']}")
        
        # If you need a DataFrame (often the next step in Data Science tasks):
        # df = pd.DataFrame(users)
        # display(df.head())
        
    else:
        print(f"Error: Status Code {response.status_code}")
        
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
from faker import Faker
import pandas as pd

fake = Faker()

# Generate a list of 10 fake users
users_data = []
for _ in range(10):
    users_data.append({
        'Name': fake.name(),
        'Email': fake.email(),
        'Address': fake.address(),
        'City': fake.city()
    })

# Convert to DataFrame
df = pd.DataFrame(users_data)
display(df)
```
