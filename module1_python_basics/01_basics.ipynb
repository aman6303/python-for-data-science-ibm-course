```python
# python operators

x = 9
y = 4 
result_add= x + y # Addition 
result_sub= x - y # Subtraction 
result_mul= x * y # Multiplication 
result_div= x / y # Division 
result_fdiv= x // y # Floor Division 
result_mod= x % y # Modulo</td>
```

```python
print(result_add)
print(result_sub)
print(result_mul)
print(result_div)
print(result_fdiv)
print(result_mod)
```

```python
# defference between del and clear

mylist = [1, 2, 3]

del mylist
print(mylist)
```

```python
# defference between del and clear

mylist = [1, 2, 3]

mylist.clear()
print(mylist). #empty list exist
```

```python
#extend vs append

mylist = [i for i in range(5)]
mylist2 = [i**2 for i in range(5)]

mylist.append(mylist2) # here we add second list as single unit 
```

```python
print(mylist) 
```

```python
mylist = [i for i in range(5)]
mylist2 = [i**2 for i in range(5)]

mylist.extend(mylist2)
```

```python
print(mylist)
```
