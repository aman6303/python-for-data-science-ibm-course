# sorted vs sort function

```python
list1 = [12, 2,3 , 4, 1]

result = sorted(list1)
print(result)
print(list1) # original lsit will not be affected
```

```python
# using sort method

result2 = list1.sort() # it return none 

print(result2)
print(list1) # original list will be affected
```

```python
# we can pass key in sorted function

mylist = ["aman", "singh", "a", "boy" ]

result = sorted(mylist, key=len)
print(result)
```

# applying function on all elements of a lsit using map

```python
list1 = [i for i in range(10)]

list_squares = list(map(lambda x: x ** 3 , list1))
print(list_squares)
```

# we can filter result using filyer function

```python
# syntax : filter(func, iterable)

nums = [1, 2, 3, 4]
evens = list(filter(lambda x:x % 2 == 0, nums))

print(evens)
```

```python
# we can combine the power of map and filter

data = [i for i in range(10)]
result = list(map(lambda x:x * 2, filter(lambda x: x > 2, data)))

print(result)
```

# use of zip function

```python
names = ['Alice', 'Bob']
ages = [25, 30]
for i,j in zip(names, ages):  # we can iterate through two list simultaneously
    print(i,j)
```

```python
print(zip(names, ages))
```

# all and any function to check multiple condition at once

```python
scores = [85, 90, 78]
all(s >= 70 for s in scores)
```

```python
scores = [85, 90, 78]
all(s >= 90 for s in scores) # any one false out put false
```

```python
# use of any

errors = [True, False, True]
if any(errors):
    print("Issues found")    # any retur true if any statement is true
```

```python
mylist = [ i for i in range(10, 21)]

if any(item > 19 for item in mylist):
    print("yes, we found an item greater than 19 in the list")
```
