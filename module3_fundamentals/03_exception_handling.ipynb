# Exception handling in python

```python
# GOOD: Using 'as e'
try:
    x = 10 / 0
except Exception as e:
    # This prints: "Error occurred: division by zero"
    print("Error occurred:", e) 

# BAD: Normal except
try:
    x = 10 / 0
except:
    # You have no idea what actually went wrong here
    print("Something went wrong")
```
